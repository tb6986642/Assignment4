```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Pudding-Film-Dialogue-Salty.csv')
```

```python
df.head()
```

## 1.Which gender has more words on average?

```python
dg=df.groupby('gender')['words'].agg(['mean']).reset_index()

plt.figure(figsize=(10,8))
sns.barplot(data=dg,x='gender',y='mean')
plt.show()
```

### Men have the most words on average

## 2.What is the quantity trend of each age? Which is more?

```python
plt.figure(figsize=(12,3))
sns.countplot(df.age)
plt.xticks(rotation=90)
plt.show()
```

### The number trend of age is now increasing and then decreasing, similar to a normal distribution. The high point is between 32 and 40, more



## 3.The percentage of gross in each year, which year is more?

```python
dt=df.groupby('year')['gross'].mean().dropna()
```

```python
y=dt.values

labels=dt.index
plt.figure(figsize=(15,15))
plt.pie(y,labels=labels, autopct = '%0.2f%%',

        shadow= False, startangle =0,labeldistance=1.2,pctdistance=0.8,

        counterclock=False,wedgeprops=dict(width=0.4,edgecolor='w'))
plt.title('gross ratio per year')
plt.show()
```

### 1967 had the most gross
